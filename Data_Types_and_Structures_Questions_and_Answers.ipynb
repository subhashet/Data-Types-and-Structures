1. What are data structures and why are they important?
   
   A data structure is a specialized format for organizing, processing, retrieving and storing data. There are several basic and advanced types of data structures, all designed to arrange data to suit a specific purpose. They are used in almost every program and software system.
2.  Explain the difference between mutable and immutable data types with    examples
    Data types in Python are categorized into mutable and immutable data types. Mutable data types are those whose values can be changed.  Example:  List, dictionaries and set.  Immutable data types are ones in which the values can’t be changed.  Example:  String and tuples.

3.  What are the main differences between lists and tuples in Python?
    
    Lists are mutable, meaning you can add, remove, or modify their elements. Tuples are immutable, meaning you cannot change them once they are created.
    Performance:
    Tuples are faster than lists because they have a fixed size, which allows Python to optimize storage and access.

    Built-in methods: Lists have more built-in methods, such as append and remove, while tuples have fewer, such as count and index.

    Use case:  Lists are better for data that changes frequently, while tuples are better for data that should remain the same.

11.  What advantages do dictionaries offer over lists for certain tasks?
